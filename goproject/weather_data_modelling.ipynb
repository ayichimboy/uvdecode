```python
def get_data(data, date_col):
    data["month"] = data[date_col].dt.month
    data['day'] = data[date_col].dt.month
    data['year'] = data[date_col].dt.year
    data['minute'] = data[date_col].dt.minute
    data['hour'] = data[date_col].dt.hour
    data['second'] = data[date_col].dt.second
    
    return data
    
```

```python
get_data(sensor_cleaned, "date-time")
```

```python
sensor_cleaned.head()
```

```python
# sort and clean the data 
df_forcast = sensor_cleaned.copy()
df_forcast = df_forcast.sort_values("date-time")
df_forcast = df_forcast.set_index("date-time")

df_forcast = df_forcast.dropna()
```
